# 감성대화 .csv로 내보내기

```python
import pandas as pd

# 1. 엑셀 파일 읽기
file_path_1 = '../aiffel/emotion_train.xlsx'  # 엑셀 파일 경로
file_path_2 = '../aiffel/emotion_val.xlsx'  # 엑셀 파일 경로

df_1 = pd.read_excel(file_path_1)
df_2 = pd.read_excel(file_path_2)
```

```python
df_1.info()
```

```python
df_2.info()
```

```python
# 2. 두 파일을 합치기 (수평으로 병합)
df = pd.concat([df_1, df_2], axis=0, ignore_index=True)

df.info()
```

```python
# 3. 필요한 열을 결합하여 새로운 열 만들기
def format_conversation(row):
    # 사람문장1, 시스템문장1, 사람문장2, 시스템문장2, 사람문장3, 시스템문장3을 결합
    conversation = []
    
    # 사람문장1과 시스템문장1
    conversation.append(row['사람문장1'])
    conversation.append(row['시스템문장1'])
    
    # 사람문장2와 시스템문장2
    conversation.append(row['사람문장2'])
    conversation.append(row['시스템문장2'])
    
    # 사람문장3과 시스템문장3 (있는 경우만 추가)
    if pd.notna(row['사람문장3']) and pd.notna(row['시스템문장3']):
        conversation.append(row['사람문장3'])
        conversation.append(row['시스템문장3'])
    
    # 각 문장을 '\n'으로 결합하여 반환
    return '\n'.join(conversation)

# 4. 각 행에 대해 `format_conversation` 함수 적용
df['conversation'] = df.apply(format_conversation, axis=1)
```

```python
df.tail(2)
```

```python
# 4. 결과를 CSV로 저장 (기존 열은 제외하고 '대화' 열만 저장)
output_path = 'emotion_dialogue.csv'  # 저장할 파일 경로
df['conversation'].to_csv(output_path, index=False, encoding='utf-8')

print("CSV 파일로 저장 완료!")
```

```python
df = pd.concat([df_01, df_02, df_03, df_04, df_05, df_06, df_07, df_08, df_09], axis=0, ignore_index=True)
df.tail(3)
```

```python
# 2. 대화 추출 및 연결하는 함수
def process_conversations(df):
    conversations = []  # 대화들을 저장할 리스트
    conversation = []   # 한 대화에 포함될 발화 모음
    last_speaker = None  # 마지막 발화자
    
    for i, row in df.iterrows():
        # 한 대화의 시작 (SENTENCEID가 1이면 새로운 대화 시작)
        if row['SENTENCEID'] == 1 and conversation:
            conversations.append('\n'.join(conversation))  # 이전 대화 끝내기
            conversation = []  # 새로운 대화 시작

        # 발화 추가 (발화자가 누구든지 간에)
        conversation.append(row['SENTENCE'])
    
    # 마지막 대화도 추가
    if conversation:
        conversations.append('\n'.join(conversation))
    
    return conversations
```

```python
# 3. 대화 리스트 가져오기
conversations = process_conversations(df)
conversations
```

```python
# 4. 결과를 새로운 데이터프레임으로 저장
result_df = pd.DataFrame({"conversation": conversations})
result_df.head(5)
```

```python
# 4. 결과를 CSV로 저장 (기존 열은 제외하고 '대화' 열만 저장)
output_path = 'korean_dialogue.csv'  # 저장할 파일 경로
result_df.to_csv(output_path, index=False, encoding='utf-8')

print("CSV 파일로 저장 완료!")
```
